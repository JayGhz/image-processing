```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Leer la imagen

```python
imagen = cv2.imread("imagen.jpg")[:,:,::-1]
```

```python
# Obtener dimensiones
alto, ancho, canales = imagen.shape
```

```python
# mostrar las dimensiones!
print(f"Dimensiones: {imagen.shape}")
```

```python
plt.imshow(imagen)
```

# Traslacin

```python
# Matriz de traslacin (mover 100px derecha y 50px abajo)
M_traslacion = np.float32([[1, 0, 100], [0, 1, 50]])
imagen_trasladada = cv2.warpAffine(imagen, M_traslacion, (ancho, alto))
plt.imshow(imagen_trasladada)
```

# Rotacin

```python
# Rotar 45 grados alrededor del centro
centro = (ancho // 2, alto // 2)
M_rotacion = cv2.getRotationMatrix2D(centro, 45, 1.0)  # (centro, ngulo, escala)
imagen_rotada = cv2.warpAffine(imagen, M_rotacion, (ancho, alto))
plt.imshow(imagen_rotada)
```

# Escala

```python
# Escalar al 150% en ambas direcciones
imagen_escalada = cv2.resize(imagen, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
plt.imshow(imagen_escalada)
```

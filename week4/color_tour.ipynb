## Colores RGB

---

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#@title Colores RGB { run: "auto" }
R = 255 #@param {type:"slider", min:0, max:255, step:1}
G = 255 #@param {type:"slider", min:0, max:255, step:1}
B = 0 #@param {type:"slider", min:0, max:255, step:1}

R1 = 225 #@param {type:"slider", min:0, max:255, step:1}
G1 = 106 #@param {type:"slider", min:0, max:255, step:1}
B1 = 0 #@param {type:"slider", min:0, max:255, step:1}
```

```python

x, y = 100, 100
m_x = x // 2
m_y = y // 2

radio = np.sqrt(m_x**2 + m_y**2)

# vamos a crear una imagen de 100x100x3 (la tercera dimensi√≥n son los colores)
imagen = np.zeros((100,100,3), dtype=np.uint8)
for i in range(100):
  for j in range(100):
    d = np.sqrt((i-m_x)**2 + (j-m_y)**2)
    t = d / radio # 0 a 1
    imagen[i, j, 0] = R + (R1 - R)*t
    imagen[i, j, 1] = G + (G1 - G)*t
    imagen[i, j, 2] = B + (B1 - B)*t

plt.figure(0)
# plt.title( "({}, {}, {})".format(R,G,B))
plt.imshow(imagen)
plt.show()
```
